# 01 — Data Collection

This notebook loads and inspects the Samsung smartphone raw dataset.

The dataset contains **612 Samsung phones** (no tablets) spanning from 2011 to 2025.
Latest phones from 2024–2025 are included with additional ML-friendly columns.

**Input:** `data/raw/SamsungPhoneData.csv`

```python
import pandas as pd
import numpy as np

print("Environment Ready!")
```

```python
df = pd.read_csv("../data/raw/SamsungPhoneData.csv")

print("Dataset Shape:", df.shape)
df.head()
```

```python
print("Columns:")
print(list(df.columns))
print(f"\nTotal columns: {len(df.columns)}")
```

```python
df.info()
```

```python
print("Missing values per column:")
df.isnull().sum()
```

```python
print("Duplicates:", df.duplicated().sum())
```

## Column Descriptions

| Column | Description |
|---|---|
| Name | Device name |
| Dimensions | Physical dimensions |
| SoC | System on Chip |
| CPU | Processor details |
| GPU | Graphics processor |
| RAM | Memory specification |
| Storage | Internal storage |
| Display | Screen specification |
| Battery | Battery specification |
| OS | Operating system |
| Camera | Camera specification |
| Launch_Year | Year of launch |
| Launch_Price | Launch price in INR |
| Series | Device series classification |
| Main_Camera_MP | Main camera megapixels |
| UltraWide_MP | Ultra-wide camera megapixels |
| Telephoto_MP | Telephoto camera megapixels |
| Front_Camera_MP | Front camera megapixels |
| OIS | Optical Image Stabilization |
| AI_Features | AI feature level |
| Waterproof | Water resistance |
| Target_Segment | Market segment |

```python
# Verify: no tablet entries
tablet_count = df["Name"].str.contains("Tab", case=False, na=False).sum()
print(f"Tablet entries: {tablet_count} (should be 0)")

# Verify: latest phones present
print(f"\nPhones from 2024+: {(df['Launch_Year'] >= 2024).sum()}")
print(f"\nLaunch Year range: {df['Launch_Year'].min()} to {df['Launch_Year'].max()}")

# Verify: no Current_Price column
print(f"\nHas Current_Price column: {'Current_Price' in df.columns} (should be False)")
```
