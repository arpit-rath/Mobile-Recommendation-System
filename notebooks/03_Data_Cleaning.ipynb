# 03 — Data Cleaning

## Objective

Clean and standardize the Samsung smartphone dataset before EDA.

The cleaning process includes:
- Missing value analysis
- Duplicate detection
- Text standardization (OIS, Waterproof, AI_Features)
- Camera MP missing value imputation

**Input:** `data/raw/SamsungPhoneData.csv`

**Output:** `data/processed/cleaned_dataset.csv`

```python
import pandas as pd
import numpy as np
import re

pd.set_option("display.max_columns", None)

print("Libraries Imported Successfully")
```

```python
df = pd.read_csv("../data/raw/SamsungPhoneData.csv")

print("Dataset Loaded Successfully")
print("Shape:", df.shape)
df.head()
```

## Missing Value Analysis

```python
print("Missing values per column:")
print(df.isnull().sum())
print(f"\nTotal missing values: {df.isnull().sum().sum()}")
```

## Duplicate Detection

```python
duplicates = df.duplicated().sum()
print(f"Duplicate rows: {duplicates}")

if duplicates > 0:
    print("Removing duplicates...")
    df = df.drop_duplicates().reset_index(drop=True)
    print(f"New shape: {df.shape}")
```

## Text Standardization

```python
# Standardize OIS and Waterproof columns
for col in ["OIS", "Waterproof"]:
    if col in df.columns:
        df[col] = df[col].astype(str).str.strip().str.title()
        print(f"Standardized {col}:")
        print(df[col].value_counts())
        print()
```

```python
# Standardize AI_Features
if "AI_Features" in df.columns:
    df["AI_Features"] = (
        df["AI_Features"]
        .astype(str)
        .str.strip()
        .replace({"None": "No", "Nan": "No", "nan": "No", "NaN": "No"})
    )
    print("Standardized AI_Features:")
    print(df["AI_Features"].value_counts())
```

## Camera MP Missing Value Handling

```python
# Fill missing camera megapixels with 0
camera_cols = ["UltraWide_MP", "Telephoto_MP"]
for col in camera_cols:
    if col in df.columns:
        missing_before = df[col].isna().sum()
        df[col] = df[col].fillna(0)
        print(f"{col}: filled {missing_before} missing values with 0")
```

## Final Validation

```python
print("Final dataset info:")
df.info()
print(f"\nFinal shape: {df.shape}")
print(f"Remaining missing values: {df.isnull().sum().sum()}")
```

## Save Cleaned Dataset

```python
output_path = "../data/processed/cleaned_dataset.csv"
df.to_csv(output_path, index=False)
print(f"Cleaned dataset saved to: {output_path}")
print(f"Shape: {df.shape}")
```
