# 04 — Exploratory Data Analysis (EDA)

## Objective

Explore the cleaned Samsung smartphone dataset to understand feature distributions,
identify relationships between variables, and derive insights for feature engineering.

**Input:** `data/processed/cleaned_dataset.csv`

**Output:** Visualizations and statistical summaries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
pd.set_option("display.max_columns", None)

print("Libraries Imported Successfully")
```

```python
df = pd.read_csv("../data/processed/cleaned_dataset.csv")
print("Dataset Shape:", df.shape)
df.head()
```

## 1. Launch Year Distribution

```python
fig, ax = plt.subplots(figsize=(12, 5))
year_counts = df["Launch_Year"].value_counts().sort_index()
year_counts.plot(kind="bar", ax=ax, color="#1a73e8", edgecolor="white")
ax.set_title("Samsung Phone Launches by Year", fontsize=14, fontweight="bold")
ax.set_xlabel("Launch Year")
ax.set_ylabel("Number of Phones")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 2. Series Distribution

```python
fig, ax = plt.subplots(figsize=(10, 5))
series_counts = df["Series"].value_counts()
colors = ["#1a73e8", "#ea4335", "#fbbc04", "#34a853", "#ff6d01", "#46bdc6", "#7b1fa2"]
series_counts.plot(kind="barh", ax=ax, color=colors[:len(series_counts)], edgecolor="white")
ax.set_title("Phones by Series", fontsize=14, fontweight="bold")
ax.set_xlabel("Count")
plt.tight_layout()
plt.show()
```

## 3. Target Segment Distribution

```python
fig, ax = plt.subplots(figsize=(8, 5))
segment_counts = df["Target_Segment"].value_counts()
segment_counts.plot(kind="pie", ax=ax, autopct="%1.1f%%", startangle=90,
                    colors=["#34a853", "#ea4335", "#1a73e8", "#fbbc04"])
ax.set_title("Target Segment Distribution", fontsize=14, fontweight="bold")
ax.set_ylabel("")
plt.tight_layout()
plt.show()
```

## 4. Launch Price Distribution

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Histogram
df["Launch_Price"].dropna().plot(kind="hist", bins=30, ax=axes[0], color="#1a73e8", edgecolor="white")
axes[0].set_title("Launch Price Distribution", fontsize=12, fontweight="bold")
axes[0].set_xlabel("Launch Price (INR)")

# Box plot by segment
df.boxplot(column="Launch_Price", by="Target_Segment", ax=axes[1])
axes[1].set_title("Launch Price by Segment", fontsize=12, fontweight="bold")
axes[1].set_xlabel("Target Segment")
axes[1].set_ylabel("Launch Price (INR)")
plt.suptitle("")

plt.tight_layout()
plt.show()
```

## 5. Camera MP Analysis

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Main Camera MP distribution
df["Main_Camera_MP"].plot(kind="hist", bins=20, ax=axes[0], color="#ea4335", edgecolor="white")
axes[0].set_title("Main Camera MP Distribution", fontsize=12, fontweight="bold")
axes[0].set_xlabel("Megapixels")

# Main Camera MP by series (2024+)
latest = df[df["Launch_Year"] >= 2024]
if len(latest) > 0:
    latest.boxplot(column="Main_Camera_MP", by="Series", ax=axes[1], rot=45)
    axes[1].set_title("Camera MP by Series (2024+)", fontsize=12, fontweight="bold")
    plt.suptitle("")

plt.tight_layout()
plt.show()
```

## 6. AI Features & Waterproof Analysis

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 5))

df["AI_Features"].value_counts().plot(kind="bar", ax=axes[0], color="#7b1fa2", edgecolor="white")
axes[0].set_title("AI Features", fontsize=12, fontweight="bold")
axes[0].tick_params(axis="x", rotation=45)

df["Waterproof"].value_counts().plot(kind="bar", ax=axes[1], color="#46bdc6", edgecolor="white")
axes[1].set_title("Waterproof", fontsize=12, fontweight="bold")
axes[1].tick_params(axis="x", rotation=0)

df["OIS"].value_counts().plot(kind="bar", ax=axes[2], color="#ff6d01", edgecolor="white")
axes[2].set_title("OIS Support", fontsize=12, fontweight="bold")
axes[2].tick_params(axis="x", rotation=0)

plt.tight_layout()
plt.show()
```

## 7. Correlation Heatmap (Numeric Features)

```python
numeric_cols = ["Launch_Year", "Launch_Price", "Main_Camera_MP", "UltraWide_MP",
                "Telephoto_MP", "Front_Camera_MP"]
numeric_df = df[numeric_cols].apply(pd.to_numeric, errors="coerce")

fig, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(numeric_df.corr(), annot=True, fmt=".2f", cmap="RdYlBu_r", ax=ax,
            linewidths=0.5, square=True)
ax.set_title("Feature Correlation Heatmap", fontsize=14, fontweight="bold")
plt.tight_layout()
plt.show()
```
