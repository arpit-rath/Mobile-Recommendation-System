# 02 — Dataset Validation & Overview

This notebook validates the Samsung smartphone dataset after data integration.

Checks performed:
- No tablet entries exist
- Latest 2024–2025 phones are present
- No `Current_Price` column
- Series, segment, and year distributions are correct

**Input:** `data/raw/SamsungPhoneData.csv`

```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/raw/SamsungPhoneData.csv")
print("Dataset Shape:", df.shape)
print("Columns:", list(df.columns))
```

## Validation Checks

```python
# Check 1: No tablets
tablet_mask = df["Name"].str.contains("Tab", case=False, na=False)
assert tablet_mask.sum() == 0, f"Found {tablet_mask.sum()} tablet entries!"
print("✅ No tablet entries found")

# Check 2: No Current_Price column
assert "Current_Price" not in df.columns, "Current_Price column still exists!"
print("✅ No Current_Price column")

# Check 3: 2024+ phones exist
new_phones = df[df["Launch_Year"] >= 2024]
assert len(new_phones) >= 29, f"Expected >=29 new phones, found {len(new_phones)}"
print(f"✅ {len(new_phones)} phones from 2024+ found")

# Check 4: Required columns present
required_cols = ["Name", "SoC", "RAM", "Storage", "Battery", "Display", "OS", "Camera",
                 "Launch_Year", "Launch_Price", "Series", "Main_Camera_MP",
                 "OIS", "AI_Features", "Waterproof", "Target_Segment"]
missing = [c for c in required_cols if c not in df.columns]
assert len(missing) == 0, f"Missing columns: {missing}"
print("✅ All required columns present")

print("\n🎉 All validation checks passed!")
```
